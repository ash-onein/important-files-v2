```python
from langdetect import detect
from transformers import pipeline
import re

# Load translation pipeline once
pipe = pipeline(
    "translation",
    model="facebook/nllb-200-distilled-600M",
    max_length=1000,
    device=0  # Ensure you're running this on a GPU; otherwise, use -1 for CPU
)

# Pre-compiled regex for sentence splitting
sentence_splitter = re.compile(r'(?<=[.!।|])\s+')

# Language map
lang_map = {
    'en': 'eng_Latn',
    'te': 'tel_Telu',
    'kn': 'kan_Knda',
    'hi': 'hin_Deva',
    'ml': 'mal_Mlym',
    'ta': 'tam_Taml',
    'bn': 'ben_Beng',
    'mr': 'mar_Deva',
    'gu': 'guj_Gujr',
    'or': 'ori_Orya',
}

def translate_text(text):
    detected_lang = detect(text)
    src_lang = lang_map.get(detected_lang, 'eng_Latn')

    if src_lang == 'eng_Latn':
        return text

    # Split and filter sentences
    sentences = [s.strip() for s in sentence_splitter.split(text) if s.strip()]

    # Batch translate instead of translating each sentence individually
    results = pipe(sentences, src_lang=src_lang, tgt_lang="eng_Latn")

    # Extract and join translated text
    translations = [res['translation_text'] for res in results]
    return ' '.join(translations)
```

# code for getting article count under each Trends

```python
from tqdm import tqdm  # optional: progress bar
from rapidfuzz import process, fuzz
from collections import defaultdict, Counter

# Prepare keywords
trend_keywords = df_trends['Trend_breakdown'].dropna().str.split(', ').explode().str.lower().unique().tolist()

# Setup translation
pipe = pipeline("translation", model="facebook/nllb-200-distilled-600M", max_length=1000, device=0)
sentence_splitter = re.compile(r'(?<=[.!।|])\s+')
lang_map = {
    'en': 'eng_Latn', 'te': 'tel_Telu', 'kn': 'kan_Knda', 'hi': 'hin_Deva',
    'ml': 'mal_Mlym', 'ta': 'tam_Taml', 'bn': 'ben_Beng', 'mr': 'mar_Deva',
    'gu': 'guj_Gujr', 'or': 'ori_Orya',
}

# Translation function
def translate_if_needed(text):
    try:
        lang = detect(text)
    except:
        lang = 'en'
    src_lang = lang_map.get(lang, 'eng_Latn')
    if src_lang == 'eng_Latn':
        return text.lower()
    sentences = [s.strip() for s in sentence_splitter.split(text) if s.strip()]
    try:
        translations = pipe(sentences, src_lang=src_lang, tgt_lang='eng_Latn')
        return ' '.join([res['translation_text'] for res in translations]).lower()
    except:
        return text.lower()

# Combine and translate
df['combined_text'] = (df['Title'].fillna('') + ' ' + df['Content'].fillna('')).astype(str)
tqdm.pandas(desc="Translating")
df['translated_text'] = df['combined_text'].progress_apply(translate_if_needed)

# Convert texts to array
texts = df['translated_text'].to_numpy()

# Match keywords to all rows using fuzzy matching
print("Matching trends to rows...")
scores = process.cdist(
    trend_keywords,
    texts,
    scorer=fuzz.partial_ratio,
    score_cutoff=70
)

# Count matches per keyword
match_counter = Counter()
row_matches = defaultdict(list)

for keyword_idx, matches in enumerate(scores):
    keyword = trend_keywords[keyword_idx]
    for row_idx, _ in matches:
        match_counter[keyword] += 1
        row_matches[row_idx].append(keyword)

# DataFrame of keyword match counts
result_df = pd.DataFrame(match_counter.items(), columns=['Keyword', 'MatchCount']).sort_values(by='MatchCount', ascending=False)

# DataFrame of matched rows
matched_row_indices = list(row_matches.keys())
df_matched = df.loc[matched_row_indices].copy()
df_matched['Matched_Keywords'] = df_matched.index.map(lambda idx: row_matches[idx])

# Show top results
print("Top keywords:\n", result_df.head(10))
print("\nTop matching rows:\n", df_matched[['Title', 'Matched_Keywords']].head(10))

# Optional: Save to CSV
result_df.to_csv('trend_match_counts.csv', index=False)
df_matched.to_csv('articles_matching_trends.csv', index=False)
```
